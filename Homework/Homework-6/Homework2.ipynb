```python
df['date'] = pd.to_datetime(df['date'])  # 형태를 바꿔줌 
```

```python
df2['date'] = pd.to_datetime(df2['date'])
```

```python
# if 문을 활용해서 연도 -> 월 -> 일 -> 시간으로 검색
# 만약 같은 값이 나온다면 세부적인 시간을 통해 xx시 00분에 라벨링 할지 xx시 30분에 라벨링할지 정함
# 모든 데이터를 쓸 수 없기 때문에 기사가 나온 시점과 제일 비슷한 시점의 주식 데이터에 라벨링 하고자 했음 -> 실패 
# ex) 기사가 10시 27분에 나왔다면 10시 30분 주식 데이터에 감성표현을 라벨링 하고자 함 

for i in range(len(df['date'])): 
  for j in range(len(df['date'])):
    if(df['date'].dt.year[i] == df2['date'].dt.year[j]):
      if(df['date'].dt.month[i] == df2['date'].dt.month[j]):
        if(df['date'].dt.date[i] == df2['date'].dt.date[j]):
          if(df['date'].dt.hour[i] == df2['date'].dt.hour[j]):
            print(df['date'].dt.date[i])
            print(df2['date'].dt.date[j])
            
```
